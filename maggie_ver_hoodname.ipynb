```python
# Fume Hood Occ Sensor

# MAGGIE'S VERSION OF THE EDGE CASES
# /Cornell University/Biotech Server/Biotech/Biotech 2nd Floor/12-2021 Controls Upgrade/Lab 203 Suite
# /Cornell University/Biotech Server/Biotech/Biotech 2nd Floor/12-2021 Controls Upgrade/Lab 209 Suite
# /Cornell University/Biotech Server/Biotech/Biotech 3rd Floor/3/2022 Lab Renovation/Lab 336
# /Cornell University/Biotech Server/Biotech/Biotech 3rd Floor/3/2022 Lab Renovation/Lab 353


def get_lab_index_location(index):
    '''
    Gets the index of the lab pointname. This is for the pointname locations 
    where they do not have it already.

    Parameters
    ----------
    index : int
        the index of the pointname on the csv

    Returns
    -------
    location of the pointname : str
        the string that has the pointname of the missing location

    '''
    # to get the location of the fume hood.
    # get the pointname index
    # find the index closest to the point name index that gets the location of the fumehood
    # join the location with the fumehood location
    # remove everything before biotech and put a hashtag at the beginning of the location


def edge_csv_parser(df):
    '''
    This function gets the edge cases of the biotech building
    Specifically the OCC and the Zone Temp
    '''
    hood_flows = df[df["Pointname"] == "FH-1 Flow"]["Point"].tolist()
    sash_pos = df[df["Pointname"] == "FH-1 Sash Position"]["Point"].tolist()
    occ_sens = df[df["Pointname"] == "Occupancy Sensor"]["Point"].tolist()
    print(sash_pos)

    occ_sens = [i for i in occ_sens if re.search(
        r"#biotech_\d+(?!a)b?_(west_|east_)?hvac/occ_sensor", i)]

    zone_temps = df[df["Pointname"] == "Zone Temp"]["Point"].tolist()
    filtered_zone_temps = []
    # #biotech_[a-zA-Z0-9]+b?_(west_|east_)?hvac/zone_temp(?!_av)
    pattern = re.compile(r"#biotech_\d+(?!a)[b]?_(west_|east_)?hvac/zone_temp(?!_av)")

    for item in zone_temps:
        if pattern.search(item):
            filtered_zone_temps.append(item)

    # Create the DataFrames
    ovals = {"Occupancy": occ_sens, "Internal Temp": filtered_zone_temps}
    ndf = pd.DataFrame(data=ovals)

    # Create another DataFrame for Flow Sensor and Sash Position
    # Assuming these lists are of the same length
    vals = {"Flow Sensor": hood_flows, "Sash": sash_pos}
    odf = pd.DataFrame(data=vals)
    return ndf, odf


edge_cases = edge_csv_parser(pointnames)[0]
edge_cases
# we dont want some of the numbers on the occ and internal. it got too much and missed
# a few bruh
# need 203b instead of 203 ( 203 one doesnt have fh anymore.... confused)
# need 245b


# dont need 351 & & 354 & 355 & 359
# biotech_lab_203b_hvac/zone_temp
```

```python
display(combined)
def get_floor(text):
    blocks = text.split("/")
    floor_block = blocks[1]
    floor = floor_block.split("_")[1]
    return floor

combined['Floor'] = [get_floor(i) for i in combined["Flow Sensor"]]
#combined

def get_lab(text):
    blocks = text.split("/")
    lab_block = blocks[3]
    lab = lab_block.split("_")[1]
    return lab

combined["Lab"] = [get_lab(i) for i in combined["Flow Sensor"]]
# replace the floor with one number/letter and for lab make it uniform
combined['Floor'] = combined['Floor'].str[:1].str.upper()
combined['Lab'] = combined['Lab'].str.upper()

# add building column 
def get_building(text):
    blocks = text.split("/")
    building_block = blocks[0]
    return building_block[1:]

combined['building'] = [get_building(i) for i in combined["Flow Sensor"]]

# add hood column (I left it as 1 for now put in manually)
combined['hood'] = "1"

# add name column UMM FIX THIS???
# def get_name(text):
#     return str(combined['building']) + "Floor_" + str(combined['Floor']) + "Lab_" + str(combined['Lab']) + ".Hood_" + "1"
# combined['name'] = [get_name(i) for i in combined]

combined
```

```python
#Rearranging the column orders :)


combined["External Temp"] = "#biotech/ground_flr_mech/building_hydronic_heating_syatems/reheat_heat_exchanger/oat"
combined["Server"] = "biotech_main"
cols = ["building","hood","Lab","Floor","Server", "Flow Sensor","Sash","Occupancy","Internal Temp","External Temp"]
combined = combined[cols]
#combined.to_csv('C:\\Users\\Rob\\Documents\\GitHub\\fume-hood-dashboard\\pointnames.csv')
combined
df = combined.copy()
floor_map = {"ground":"Ground", "1st":"First","2nd":"Second", "3rd":"Third","4th":"Fourth"}
df["FloorNum"] = df["Floor"].map(floor_map)
display(df)
#Biotech.floor_num.lab_num.hood_num.{pointType}
#4 Point Types:[energy.occ,energy.unocc,sashOpenTime.occ, sashOpenTime.unocc]

def point_maker(lab,floor,hood):
    nstr = "Biotech." + str(floor) + "Floor" + ".Lab_" +  str(lab)   + ".Hood_" + str(hood) + "."
    nlist = [nstr+"energy.occ",nstr+"energy.unocc",nstr+"sashOpenTime.occ",nstr+"sashOpenTime.unocc"]
    return nlist



#display(df)

# test_case = point_maker(141,1,1)
# test_case

def make_all_points(df):
    energy_occ = []
    energy_unocc = []
    sashOpenTime_occ = []
    sashOpenTime_unocc = []
    for i in range(len(df["FloorNum"])):
        response = point_maker(df["Lab"][i],df["FloorNum"][i],1)
        energy_occ.append(response[0])
        energy_unocc.append(response[1])
        sashOpenTime_occ.append(response[2])
        sashOpenTime_unocc.append(response[3])
    df["Energy Occupied"] = energy_occ
    df["Energy Unoccupied"] = energy_unocc
    df["Sash Open Time Occupied"] = sashOpenTime_occ
    df["Sash Open Time Unoccupied"] = sashOpenTime_unocc
    return df

ndf = make_all_points(df)
display(ndf)






# ndf.to_csv("biotech_point_names.csv")
```
